```python
import numpy as np
import matplotlib.pyplot as plt
```

### Load data

```python
eps_target_list = np.linspace(start=0.5,stop=4,num=15)

mse_csgm_n100_d20_g03 = np.load('results/mse_csgm_n100_d20_g03.npy')
mse_csgm_n100_d20_g05 = np.load('results/mse_csgm_n100_d20_g05.npy')
mse_csgm_n100_d20_g10 = np.load('results/mse_csgm_n100_d20_g10.npy')

mse_sigm_n100_d20_g03 = np.load('results/mse_sigm_n100_d20_g03.npy')
mse_sigm_n100_d20_g05 = np.load('results/mse_sigm_n100_d20_g05.npy')
mse_sigm_n100_d20_g10 = np.load('results/mse_sigm_n100_d20_g10.npy')
```

```python
mse_csgm_n1000_d100_g03 = np.load('results/mse_csgm_n1000_d100_g03.npy')
mse_csgm_n1000_d100_g05 = np.load('results/mse_csgm_n1000_d100_g05.npy')
mse_csgm_n1000_d100_g10 = np.load('results/mse_csgm_n1000_d100_g10.npy')
mse_sigm_n1000_d100_g03 = np.load('results/mse_sigm_n1000_d100_g03.npy')
mse_sigm_n1000_d100_g05 = np.load('results/mse_sigm_n1000_d100_g05.npy')
mse_sigm_n1000_d100_g10 = np.load('results/mse_sigm_n1000_d100_g10.npy')
```

### Plot Figure for $n=100$

```python
plt.figure(figsize=(5,5))
# CSGM
plt.plot(eps_target_list,mse_csgm_n100_d20_g03,
      linestyle='--',label = rf'CSGM ($\gamma$=0.3)',color='cyan',ms=4,
      marker='o',linewidth=1)
plt.plot(eps_target_list,mse_csgm_n100_d20_g05,
      linestyle='--',label = rf'CSGM ($\gamma$=0.5)',color='dodgerblue',ms=4,
      marker='o',linewidth=1)
plt.plot(eps_target_list,mse_csgm_n100_d20_g10,
      linestyle='--',label = rf'CSGM ($\gamma$=1.0)',color='navy',ms=4,
      marker='o',linewidth=1)
# SIGM
plt.plot(eps_target_list,mse_sigm_n100_d20_g03,
      label = rf'SIGM ($\gamma$=0.3)',color='gold',ms=5,
      marker='v',linewidth=1)
plt.plot(eps_target_list,mse_sigm_n100_d20_g05,
      label = rf'SIGM ($\gamma$=0.5)',color='orange',ms=5,
      marker='v',linewidth=1)
plt.plot(eps_target_list,mse_sigm_n100_d20_g10,
      label = rf'SIGM ($\gamma$=1.0)',color='red',ms=5,
      marker='v',linewidth=1)
plt.legend(fontsize=10)
plt.xlabel(r'Privacy ($\varepsilon$)', fontsize=16)
plt.text(0.0, 0.35, 'MSE', rotation=90,fontsize=16, va='center')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.yscale('log')
plt.grid(alpha=0.2,which='both')
plt.show()
```
